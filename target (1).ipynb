custom data  YAML  

```python
!pip install PyYAML


import yaml
data = { 'train' :'/content/data/train',
         'val' : '/content/data/valid',
         'test': '/content/data/test',
         'names':['trash'],
         'nc': 1}

with open('/content/data/customdata.yaml', 'w') as f:
  yaml.dump(data, f)

with open('/content/data/customdata.yaml', 'r') as f:
  taco_yaml = yaml.safe_load(f)
  display('customdata.yaml')
```

```python
from ultralytics import YOLO

# load a pre-trained model
model = YOLO("yolov8n-seg.pt")

# train the model
model.train(data="/content/data/customdata.yaml", epochs=100, imgsz=640)

# predict with the model
results = model.predict(source="test.jpg")
```
